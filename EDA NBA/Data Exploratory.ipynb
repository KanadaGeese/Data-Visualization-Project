```python
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import jupyprint as jp
```

```python
# mean
def mean(s):
    return np.mean(s)

# standard deviation
def std(s):
    return np.std(s)

# variance
def var(s):
    return np.var(s)

# covariance
def cov(x,y):
    return np.cov(x,y, bias = True)[0,1]

# correlation
def corr(x,y):
    return np.corrcoef(x,y)[0,1]

# skewness
def skew(s):
    return st.skew(s)

# kurtosis
def kurt(s):
    return st.kurtosis(s)

def stats(s):
    print(f'nobs = {len(s)}, minmax = {min(s), max(s)}, mean = {mean(s)}, std = {std(s)}, skewness = {skew(s)}, kurtosis = {kurt(s)}')
```
